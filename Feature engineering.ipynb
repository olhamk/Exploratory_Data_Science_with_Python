```python
#2. engineer any 2 new features for loans_2007.csv data 

#let's see if we have any missing values in the data 
null_counts = loans2007_data.isnull().sum()
print(null_counts)
```

```python
#1. "revol_util" has 50 missing values so we can try to fill those with the "mean"
mean_value=loans2007_data['revol_util'].mean() #mean value of "revol_util"
print(mean_value)
loans2007_data['revol_util']=loans2007_data['revol_util'].fillna(mean_value) #filling in missing values with the mean value
print(loans2007_data['revol_util'].isnull().sum()) #checking the missing values once again 
```

```python
#2. Transforming year of employment (emp_length) into categories


loans2007_data['emp_length'].unique() #list of all the unique values in 'emp_length' column
```

```python
#defining the bins as 0 - 1, 1 - 4, 4 - 7, 7 - 10 
bins = [0, 1, 4, 7, 10]
```

```python
#variable names for the employment_duration

group_names = ['< 1 years', '1 - 4 years', '4 - 7 years', '> 7 years']
```

```python
loans2007_data['employment_duration'] = pd.cut(loans2007_data['emp_length'], bins, labels=group_names)
```

```python
loans2007_data.sample(10)
```

```python
loans2007_data['employment_duration'].unique() #distinct values in 'employment_duration' column 
```

```python
print(loans2007_data['employment_duration'].isnull().sum()) #sum of missing values 
```

```python
# 0 values are treated as missing values (NaN) in the new column, so let's replace them with 'unemployed' category 

loans2007_data['employment_duration'] = loans2007_data['employment_duration'].cat.add_categories('unemployed')
loans2007_data['employment_duration'].fillna('unemployed', inplace = True)
```

```python
loans2007_data['employment_duration'].unique()
```

```python
print(loans2007_data['employment_duration'].isnull().sum()) #sum of missing values after replacement 
```
